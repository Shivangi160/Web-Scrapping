```
import requests
from bs4 import BeautifulSoup
```

```
url = "https://www.amazon.in/Test-Exclusive-544/product-reviews/B077PWK5BY/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```
page = requests.get(url)
page
```

```
soup = BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```
names = soup.select('span.a-profile-name')[2:]
names
```

```
title = soup.select('a.review-title span')
title
```

```
dates = soup.select('span.review-date')[2:]
dates
```

```
ratings = soup.select('span.a-icon-alt')[3:]
ratings
```

```
reviews = soup.select('span.review-text-content span')
reviews
```

```
cust_name = []
cust_rate = []
rev_title = []
rev_date = []
rev_body = []
for i in range(10):
  cust_name.append(names[i].get_text())
  cust_rate.append(ratings[i].get_text())
  rev_title.append(title[i].get_text())
  rev_date.append(dates[i].get_text())
  rev_body.append(reviews[i].get_text().replace("\n",""))
```

```
import pandas as pd
df = pd.DataFrame()
df['Customer Name'] = cust_name
df['Date'] = rev_date
df['Title'] = rev_title
df['Ratings'] = cust_rate
df['Reviews'] = rev_body
df
```

```
df.to_csv('reviews.csv')
```
